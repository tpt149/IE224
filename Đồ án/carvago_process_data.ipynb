# Xử lý Warranty until

```python
df['Warranty until'].unique()
```

```python
make, value = [], []
for i in range(len(df)):
    if df['First registration'][i] and len(df['Warranty until'][i]):
        make.append(df['Make'][i])
        value.append(int(df['Warranty until'][i].split('/')[1]) - df['First registration'][i])
```

```python
d = {}
for name in pd.Series(make).unique():
    l = []
    for i in range(len(make)):
        if make[i] == name:
            l.append(value[i])
    d[name] = l
```

```python
for i in range(len(df)):
    if df['Warranty until'][i]:
        df['Warranty until'][i] = int(df['Warranty until'][i].split('/')[1]) - df['First registration'][i]
    else:
        if df['Make'][i] in d:
            df['Warranty until'][i] = int(np.mean(d[df['Make'][i]]))
        else: df['Warranty until'][i] = np.nan
```

```python
df['Warranty until'] = df['Warranty until'].astype('float')
df['Warranty until'].unique()
```

# Loại bỏ Outlier

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)

# Tính IQR
IQR = Q3 - Q1

# Xác định giá trị ngoại lệ
lower_extreme = Q1 - 1.5 * IQR
upper_extreme = Q3 + 1.5 * IQR

df = df[(df['Price'] >= lower_extreme) & (df['Price'] <= upper_extreme)]
```
